```python
# Importando as dependências necessárias
from PIL import Image # Importando o PIL
import cv2 # Importando o cv2
import numpy as np # Importando o numpy
from google.colab.patches import cv2_imshow # Importando o cv2
```

```python
import dlib # Esse comando irá importar a biblioteca de software Dlib
```

```python
# Esse comando irá carregar o detector de faces da Dlib que foi importada a cima. 
detector_face = dlib.get_frontal_face_detector()

# Esse comando será utilizado para ler a imagem com faces escolhida.
imagem = cv2.imread('/content/imagem/reacao.png')

# Chama o detector e irá passar a imagem com os seus devidos argumentos.
deteccoes = detector_face(imagem, 1)

for face in deteccoes: # Aqui passaremos um for para detectar cada face
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom() # Aqui será pego as cordenadas da imagem
  
  # Aqui será criado um cv rectangle na imagem passada a cima, criando todas as cordenadas especificadas
  # Será passado também as cores do retângulo e o seu tamanho.
  cv2.rectangle(imagem, (l,t), (r, b), (1, 1, 1 ), 10)  
  
# Aqui irá mostrar o resultado final da imagem. 
cv2_imshow(imagem)
```

```python
# Aqui iremos carregar o detector de marcos faciais da DLIB.
detector_pontos = dlib.shape_predictor('/content/shape_predictor_68_face_landmarks.dat')
```

```python
#Pegando a mesma foto utilizada antes e redefinindo-a como "imagemnova" e lendo a imagem.
imagemnova = cv2.imread('/content/imagem/reacao.png')

#Aqui iremos detectar os retângulos que serão enconstrados na imagem, logo após será armazenado no vetor "deteccoes"
deteccoes = detector_face(imagemnova, 1)
for face in deteccoes: # Será feito a busca e para cada face detectada será identificado os marcos faciais que serão armazenados no vetor "pontos"
  pontos = detector_pontos(imagemnova, face) # Vetor pontos que receberá uma imagem e um retângulo como parâmetros. Buscará as informações nessa imagem e na face que foi passada. 

  # Aqui será criado e pintado todos os marcos faciais (boca, olhos, nariz, sonbrancelha e rosto)
  # Logo abaixo será passado as cordenadas para que o programa possa desenhar os pontos faciais.
  for ponto in pontos.parts():
    cv2.circle(imagemnova, (ponto.x, ponto.y), 2, (0,255,0), 1)
 
  # Aqui iremos pintar o retângulo ma imagemnova, logo após identificar os pontos, para não bugar o programa, pois pode acontecer dele não identificar uma parte da face.
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(imagemnova, (l,t), (r, b), (500,255,255), 4)

# Irá imprimir a imagem com o resultado final. 
cv2_imshow(imagemnova)
```
